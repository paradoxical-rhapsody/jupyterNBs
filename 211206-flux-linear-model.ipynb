```julia
using Random
using Flux

rng = MersenneTwister(2021) ;
```

```julia
n = 100 ;
p = 10 ;
x = rand(rng, n, p) ;
y = rand(rng, n) ;

# OLS estimator
β_ols = inv(x'*x) * x' * y;
```

```julia
# Define and initialize the model we want to train
loss(x, y) = sum((y - x * β).^2)

β = rand(rng, p) ;
opt = Descent(0.01)
ps = params(β) ;
Flux.Optimise.update!( opt, ps, gradient(()->loss(x, y), ps) )

@show β
```
